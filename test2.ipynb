```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
from scipy.sparse import csr_matrix
```

```python
csv_path = 'Resources/SpotifyFeatures.csv'
df = pd.read_csv(csv_path)

df.head()
```

```python
df.drop_duplicates()
df.dropna()
df = df.drop_duplicates(subset=['track_id'])
```

```python
df['genre'].unique()
```

```python
repl_dict={"'":""}
df["genre"]= df["genre"].replace(repl_dict,regex=True)
repl_dict={"’":""}
df["genre"]= df["genre"].replace(repl_dict,regex=True)
```

```python
df1 = df.drop(columns = ['instrumentalness', 'key', 'mode', 'duration_ms', 'time_signature', 'track_id', 'artist_name'])
df1.head()
```

```python
df1.genre[df1.genre == 'Childrens Music'] = 1
df1.genre[df1.genre == 'Anime'] = 2
df1.genre[df1.genre == 'Comedy'] = 3
df1.genre[df1.genre == 'Soundtrack'] = 4
df1.genre[df1.genre == 'Movie'] = 5
df1.genre[df1.genre == 'Pop'] = 6
df1.genre[df1.genre == 'R&B'] = 7
df1.genre[df1.genre == 'Hip-Hop'] = 8
df1.genre[df1.genre == 'Rap'] = 9
df1.genre[df1.genre == 'Reggae'] = 10
df1.genre[df1.genre == 'Reggaeton'] = 11
df1.genre[df1.genre == 'Ska'] = 12
df1.genre[df1.genre == 'Soul'] = 13
df1.genre[df1.genre == 'Blues'] = 14
df1.genre[df1.genre == 'Rock'] = 15
df1.genre[df1.genre == 'Folk'] = 16
df1.genre[df1.genre == 'Country'] = 17
df1.genre[df1.genre == 'Alternative'] = 18
df1.genre[df1.genre == 'Indie'] = 19
df1.genre[df1.genre == 'Electronic'] = 20
df1.genre[df1.genre == 'Dance'] = 21
df1.genre[df1.genre == 'World'] = 22
df1.genre[df1.genre == 'Opera'] = 23
df1.genre[df1.genre == 'A Capella'] = 24
df1.genre[df1.genre == 'Jazz'] = 25
df1.genre[df1.genre == 'Classical'] = 26
```

```python
df1 = df1.set_index('track_name')
df1.head()
```

```python
df1['genre'] = df1['genre'].astype(int)
```

```python
model_knn = NearestNeighbors(algorithm= 'kd_tree',n_neighbors= 7)
```

```python
mat_songs = csr_matrix(df1.values)
```

```python
model_knn.fit(mat_songs)
```

```python
def recommend(idx, model, number_of_recommendations= 7):
    query = df1.loc[idx].to_numpy().reshape(1,-1)
    print('Searching for recommendations...')
    distances, indices = model.kneighbors(query,n_neighbors = number_of_recommendations)
    
    for i in indices:
        print(df[['track_name','artist_name']].loc[i].where(df['track_name']!=idx).dropna())
```

```python
name = input('Enter song title: ')
print('Search results: ')
print(df[['artist_name','track_name']].where(df['track_name'] == name).dropna())

ind = int(input('Enter the index value of the required song: '))
idx = df['track_name'].loc[ind]

song = df['track_name'].loc[ind]
artists = df['artist_name'].loc[ind]

print('Song selected is ', song, 'by', artists)

nor = int(input('Enter number of recommendations: '))

recommend(idx, model_knn, nor)
```
